# Индивидуальное $№2$.

## Есть кривая, заданная параметрически с помощью двух функций $x(t)$ и $y(t)$, где $t$ - параметр, изменяющийся в интервале $[a, b]$. Наша задача состоит в вычислении длины этой кривой на заданном интервале.

```python
# импортируем библиотеки
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
# Заданные функции x(t) и y(t)
def x(t):
    return np.cos(t)

def y(t):
    return np.sin(t)
```

```python
# Интервал [a, b]
a = 0
b = 2*np.pi
```

```python
# Вычисление производных dx/dt и dy/dt
def dx_dt(t):
    return -np.sin(t)

def dy_dt(t):
    return np.cos(t)
```

```python
# Вычисление длины кривой
curve_length, error = integrate.quad(lambda t: np.sqrt((dx_dt(t))**2 + (dy_dt(t))**2), a, b)
print("Длина кривой на интервале [{}, {}]: {}".format(a, b, curve_length))
```

```python
# Построение графика кривой линии
t = np.linspace(a, b, 100)
plt.plot(x(t), y(t), 'b', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Кривая на плоскости')
plt.grid(True)
plt.show()
```
